**Exploratory Data Analysis**

```python
# Get the total Sales and total Profit
total_sales = df['sales'].sum()
total_profit = df['profit'].sum()

# Print the results
print(f"Total Sales: R{total_sales:,.2f}")
print(f"Total Profit: R{total_profit:,.2f}")
```

```python
# Group the data by 'region'
# Calculate the total sales for each region
# Then, sort the results from highest to lowest sales
sales_by_region = df.groupby('region')['sales'].sum().sort_values(ascending=False)
print(sales_by_region)
```

**Sales by Region Bar Chart**

```python
# Import library
import matplotlib.pyplot as plt

# Plot the chart
plt.figure(figsize=(10,6))
ax = sales_by_region.plot(kind='bar', color='orchid', edgecolor='lightpink', linewidth=2)
plt.title('Sales by Region', fontsize=18, fontweight='bold', color='purple')
plt.xlabel('Region', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**Sales by Profit Bar Chart**

```python
# Group the data by 'category'
# Calculate the total sales and total profit for each category
# Then sort the categories by total sales in descending order 
category_summary = df.groupby('category')[['sales', 'profit']].sum().sort_values(by='sales', ascending=False)

# Print the results
print(category_summary)
```

```python
# Plot
colors = ['orchid', 'pink']
category_summary.plot(kind='bar', figsize=(8,5), color=colors, edgecolor='lightpink', linewidth=2)
plt.title('Sales and Profit by Category', fontsize=16, fontweight='bold', color='purple')
plt.ylabel('Amount')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**Top 10 Products by Sales Bar Chart**

```python
# Group the data by 'product_name'
# Calculate the total sales for each product
# Sort the products by sales in descending order 
# Select the top 10 products with the highest sales
top_products = df.groupby('product_name')['sales'].sum().sort_values(ascending=False).head(10)

# Print the top 10 products and their total sales
print(top_products)

# Plot a horizontal bar plot for the top 10 products by sales
top_products.plot(kind='barh', figsize=(8,5), color='orchid', edgecolor='pink', linewidth=2)

# Add a title 
plt.title('Top 10 Products by Sales', fontsize=16, fontweight='bold', color='purple')

# Label the x-axis as 'Sales' 
plt.xlabel('Sales')

plt.gca().invert_yaxis()

plt.tight_layout()

# Show the plot
plt.show()
```

**Sales Distribution by Segment Pie Chart**

```python
# Import library
import matplotlib.pyplot as plt

# Group the sales data by 'Segment' and calculate the total sales for each segment
segment_sales = df.groupby('segment')['sales'].sum()

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(segment_sales, labels=segment_sales.index, autopct='%1.1f%%', colors=['orchid','pink','plum'], shadow=True, textprops={'fontsize':14}, wedgeprops={'edgecolor':'white'}, startangle=140)
plt.title('Sales Distribution by Segment', fontsize=16, fontweight='bold', color='purple')
plt.show()
```

**Correlation of Numerical Features Heatmap**

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Keep only numeric columns for correlation
num_cols = ['sales', 'quantity', 'discount', 'profit', 'shipping_cost']
corr_matrix = df[num_cols].corr()

# Plot the heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='magma', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features', fontsize=16, fontweight='bold', color='purple')
plt.show()
```

**Monthly Sales Trend Line Chart**

```python
# Ensure 'order_date' is datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Create a new column for Year-Month
df['year_month'] = df['order_date'].dt.to_period('M')

# Group by Year-Month and calculate total sales
monthly_sales = df.groupby('year_month')['sales'].sum().reset_index()

# Convert Period to datetime for plotting
monthly_sales['year_month'] = monthly_sales['year_month'].dt.to_timestamp()
```

```python
# Import library
import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['year_month'], monthly_sales['sales'], marker='o', color='orchid', linewidth=2, markerfacecolor='pink')
plt.title('Monthly Sales Trend', fontsize=16, fontweight='bold', color='purple')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45, ha='right')

for x, y in zip(monthly_sales['year_month'], monthly_sales['sales']):
    plt.text(x, y, f'{int(y):,}', ha='center', va='bottom', fontsize=6)



plt.show()
```

**Profit Distribution by Category Boxplot**

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(8, 5))
sns.boxplot(x='category', y='profit', data=df, palette='pastel')
plt.title('Profit Distribution by Category', fontsize=16, fontweight='bold', color='purple')
plt.xlabel('Category', fontsize=14)
plt.ylabel('Profit', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

**Profit vs Sales Scatter Plot by Category**

```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sales', y='profit', hue='category', alpha=0.6)
plt.title('Profit vs Sales Scatter Plot by Category', fontsize=16, fontweight='bold', color='purple')
plt.xlabel('Sales', fontsize=12)
plt.ylabel('Profit', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Sales by Category and Segment (Stacked Bar Chart)**

```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Grouping the data
stacked_data = df.groupby(['category', 'segment'])['sales'].sum().unstack().fillna(0)

# Plot
stacked_data.plot(kind='bar', stacked=True, figsize=(10,6), colormap='Pastel1')

plt.title('Sales by Category and Segment (Stacked Bar Chart)', fontsize=16, fontweight='bold', color='purple')
plt.xlabel('Category',  fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.legend(title='Segment')
plt.xticks(rotation=0)
plt.grid(True, color='gray', linestyle='--', linewidth=0.7)
plt.tight_layout()
plt.show()
```

**Profit by Region and Category (Heatmap)**

```python
# Group the data
heatmap_data = df.groupby(['region', 'category'])['profit'].sum().unstack().fillna(0)

# Plot heatmap
plt.figure(figsize=(12, 7))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="magma", linewidths=0.5)

plt.title('Profit by Region and Category (Heatmap)', fontsize=16, fontweight='bold', color='purple')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Region', fontsize=12)
plt.tight_layout()
plt.show()
```
